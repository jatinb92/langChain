# 2. Recursive Character Text Splitting


> Start with big separators → "\n\n" (paragraphs).

> If a chunk is still too big, use smaller separators → "\n" (lines), then " " (spaces).

> If still too large, force split by characters.


```python
# Original Text
#       │
#       ▼
# Split by Paragraphs ( \n\n )
#  ├── Chunk A (Paragraph 1)
#  │       │
#  │       ▼
#  │   Split by Sentences ( . )
#  │       ├── A1 (Sentence 1)
#  │       └── A2 (Sentence 2)
#  │
#  └── Chunk B (Paragraph 2)
#          │
#          ▼
#      Split by Sentences ( . )
#          ├── B1 (Sentence 1)
#          └── B2 (Sentence 2)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size = 200,
    chunk_overlap=0
)

result = splitter.split_documents(text_docs)
```

# 3. Document Based Spliitter

eg splitting python code 

```python
from langchain.text_splitter import Language
```

```python
code = """ 

import random

# A dictionary mapping grade ranges to their letter equivalents.
GRADE_MAP = {
    range(90, 101): "A",
    range(80, 90): "B",
    range(70, 80): "C",
    range(60, 70): "D",
    range(0, 60): "F"
}

def get_letter_grade(score):

    for score_range, letter in GRADE_MAP.items():
        if score in score_range:
            return letter
    return "Invalid Score"

def generate_student_data(num_students=50):
    
    print("Generating simulated student data...")
    data = []
    for i in range(1, num_students + 1):
        score = random.randint(50, 100)
        data.append({
            "name": f"Student {i:02d}",
            "id": f"S{1000 + i}",
            "score": score
        })
    return data

"""
```

```python
splitter = RecursiveCharacterTextSplitter.from_language(
    language = Language.PYTHON,
    chunk_size = 300,
    chunk_overlap=0
)

result = splitter.split_text(code)
```

# 4. Experimantal: Semantic Based Searching

```python
from langchain_experimental.text_splitter import SemanticChunker, calculate_cosine_distances
from langchain_openai.embeddings import OpenAIEmbeddings
```

```python
sample = """
Farmers were working hard in the fields, preparing the soil and planting seeds for the next season. The sun was bright, and the air smelled of earth and fresh grass. The Indian Premier League (IPL) is the biggest cricket league in the world. People all over the world watch the matches and cheer for their favourite teams.


Terrorism is a big danger to peace and safety. It causes harm to people and creates fear in cities and villages. When such attacks happen, they leave behind pain and sadness. To fight terrorism, we need strong laws, alert security forces, and support from people who care about peace and safety.
"""
```

```python
api_key = 'your_api_key'
```

```python
text_splitter = SemanticChunker(
    OpenAIEmbeddings(api_key=api_key),
    breakpoint_threshold_type= 'standard_deviation', # others include 'percentile', 'interquartile', 'gradient'
    breakpoint_threshold_amount=1 # we need to play with threshold amount to get correct value
    
)

result = text_splitter.create_documents([sample])
```

```python
print(len(result))

# print(result)

for res in result:
    print(res.page_content, '\n\n' )
```
