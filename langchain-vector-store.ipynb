# Using FAISS Vector Store

```python
import faiss
from langchain_community.docstore.in_memory import InMemoryDocstore
```

```python
embeddings = OpenAIEmbeddings(api_key=api_key)
index = faiss.IndexFlatL2(len(embeddings.embed_query("hello world")))

faiss_vector_store = FAISS(
    embedding_function=embeddings,
    docstore=InMemoryDocstore(),
    index=index,
    index_to_docstore_id={},
)
```

```python
faiss_vector_store.add_documents(naruto_docs)
```

# FAISS Retrieve document

```python
# not supported at faiss
# faiss_vector_store.get(include=['metadatas']), # or embeddings, documents
```

# FAISS Vector search

```python
results = faiss_vector_store.similarity_search_with_score(
    query="name the copy ninja", 
    k=5, # no. of desired results
    
)

for res in results:
    print(res, "\n")
```
