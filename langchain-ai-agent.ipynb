```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun
import requests
```

```python
search_tool = DuckDuckGoSearchRun()

@tool
def get_weather_data(city: str):
    '''
    This function returns the weather of the given city.
    '''

    weather_api_key = 'your_api_key'

    url=f'https://api.weatherstack.com/current?access_key={weather_api_key}&query={city}'

    response = requests.get(url)

    return response.json()
# result = search_tool.invoke('top 5 new India today')

# print(result)
```

```python
llm = ChatOpenAI(api_key=api_key)

# result = llm.invoke('Hello')

# print(result)
```

```python
from langchain.agents import create_react_agent, AgentExecutor
from langchain import hub
```

```python
# Step 2 : Pull ReAct agent prompt
#prompt = hub.pull('hwchase17/react')

prompt_react_chat = hub.pull('hwchase17/react-chat')
```

```python
# Step 3 : Create a ReAct agent

agent = create_react_agent(
    llm = llm,
    tools=[search_tool, get_weather_data],
    prompt=prompt_react_chat
)
```

```python
# Step 4: Wrap it with AgentExecutor

agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool, get_weather_data],
    verbose=True
)
```

```python
# Step 5: Invoke

# with prompt_react_chat
response = agent_executor.invoke({"input": "population of Faridabad and its weather", 'chat_history': ''})


# with prompt - react
#response = agent_executor.invoke({"input": "population of Faridabad and its weather"})

print(response)

# every prompt has its own pre-requiste
```
